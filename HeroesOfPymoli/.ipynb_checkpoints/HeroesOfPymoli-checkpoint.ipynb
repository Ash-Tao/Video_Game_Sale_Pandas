## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
UniqueItems=len(purchase_data["Item ID"].unique())
AveragePrice = purchase_data["Price"].mean().round(2)
Purchases = purchase_data["Purchase ID"].count()
TotalRevenue = purchase_data["Price"].sum().round(2)

PurchasingAnalysis_df = pd.DataFrame(
    {"Number of Unique Items": [UniqueItems],
     "Average Price":AveragePrice,
     "Number of Purchases":Purchases,
     "Total Revenue":TotalRevenue})

PurchasingAnalysis_df["Average Price"] = PurchasingAnalysis_df["Average Price"].map("${:,.2f}".format)
PurchasingAnalysis_df["Total Revenue"] = PurchasingAnalysis_df["Total Revenue"].map("${:,.2f}".format)

PurchasingAnalysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
GroudByAge= pd.DataFrame(Gender_df.groupby("Age")["Purchase ID"].count()).rename(columns={"Purchase ID":"Total Count"}).reset_index()
GroudByAge
```

```python
bins=[0,9,14,19,24,29,34,39,50]
group_name = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","+40"]
GroudByAge["Age Summary"] = pd.cut(GroudByAge["Age"], bins, labels=group_name, include_lowest=True)
GroudByAge[["Total Count","Age Summary"]]
```
